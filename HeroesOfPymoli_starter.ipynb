```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_df = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchase_df)
purchase_df
```

## Player Count

* Display the total number of players

```python
player_count = purchase_df['SN'].nunique()
player_count_df = pd.DataFrame({'Total player Count':[player_count]})
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
no_dup_items = purchase_df.drop_duplicates(['Item ID'], keep = 'last')
#counts items by unique ID
total_unique = len(no_dup_items)
#finds the number of total purchases by counting occurances of price
total_pur = purchase_df['Price'].count()
#calculates total revenue for table by summing occurance of price and below calc
total_rev = round(purchase_df['Price'].sum(),2)
#calculates total_rev
avg_price = round(total_rev/total_pur, 2)

#creates Purchase Analysis DataFrame
purchase_analysis = pd.DataFrame([{
    
    "Number of Unique Items": total_unique,
    'Average Purchase Price': avg_price,
    'Total Purchases': total_pur,
    'Total Revenue': total_rev
}])
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
new_df = purchase_df[['SN','Age','Gender']]
new_df = new_df.groupby(['Gender']).nunique(dropna = True)
new_df = new_df[['SN']]
new_df['Percentage of Players']= round(100*(new_df/new_df.sum()),2)
new_df['Percentage of Players'] = new_df['Percentage of Players'].map("{}%".format)
new_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pur_count_by_gen = pd.DataFrame(purchase_df.groupby('Gender')['Gender'].count())
# sums price by gender
total_pur_by_gen = pd.DataFrame(purchase_df.groupby('Gender')['Price'].sum())
pur_analysis_gen = pd.merge(pur_count_by_gen, total_pur_by_gen, left_index = True, right_index = True)
pur_analysis_gen.rename(columns = {'Gender': 'Purchases count', 'Price':'Total Purchase Value'}, inplace=True)
#adding colunn
pur_analysis_gen['Average Purchase Price'] = round(pur_analysis_gen['Total Purchase Value']/pur_analysis_gen['Purchases count'],2)
pur_analysis_gen
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [-1, 9, 14, 19, 24, 29, 34, 39, 75]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34','35-39', '40=<']
age_dem = purchase_df
age_dem["Age Range"]= pd.cut(age_dem['Age'],bins,labels=group_names)
age_dem.head()
age_dem = age_dem.groupby('Age Range')
demdf = pd.DataFrame(age_dem['Age'].count())
demdf
result = demdf.divide(other = player_count)*100
demdf['Percentage']=result
demdf['Percentage']=demdf['Percentage'].map("{:.2f}%".format)
demdf
```

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
age_purchase = pd.DataFrame(age_dem['Purchase ID'].count())
age_purchase = age_purchase.rename(columns={'Purchase ID':'Purchase Count'})
age_purchase['Average Purchase Price']=age_dem['Price'].mean().map("${:.2f}".format)
age_purchase['Total Purchase Value']=age_dem['Price'].sum().map("${:.2f}".format)
age_purchase['Avg Total Purchase per Person']=(age_dem['Price'].sum()/(age_dem['SN'].nunique())).map("${:.2f}".format)

age_purchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#purchase_df.groupby(['SN']).nunique(dropna = True)
#groupby columns
purchase_amt_by_SN = pd.DataFrame(purchase_df.groupby('SN')['Price'].sum())
num_purchase_by_SN = pd.DataFrame(purchase_df.groupby('SN')['Price'].count())
avg_purchase_by_SN = pd.DataFrame(purchase_df.groupby('SN')['Price'].mean())
#merge and sort
merged_top5 = pd.merge(purchase_amt_by_SN, num_purchase_by_SN, left_index = True, right_index = True).merge(avg_purchase_by_SN, left_index=True, right_index=True)
merged_top5.rename(columns = {'Price_x': 'Total Purchase Value', 'Price_y':'Purchase Count', 'Price':'Average Purchase Price'} , inplace = True)
merged_top5.sort_values('Total Purchase Value', ascending = False, inplace=True)
merged_top5 = merged_top5.head()
round(merged_top5.head(),2)                                                                                                            
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

mpi = purchase_df.groupby(['Item ID', 'Item Name'])
summary_profitable=pd.DataFrame(mpi['Price'].count())
summary_profitable['Item Price']=mpi['Price'].max()
summary_profitable['Total Purchase Value']=mpi['Price'].sum()
final_summary_profitable=summary_profitable.sort_values('Total Purchase Value',ascending=False)
final_summary_profitable = final_summary_profitable.rename(columns={'Price':'Purchase Count'})
final_summary_most_popular = summary_profitable.sort_values('Price',ascending=False)
final_summary_most_popular = final_summary_most_popular.rename(columns={'Price':'Purchase Count'})


final_summary_most_popular['Item Price']=final_summary_most_popular['Item Price'].map("${:.2f}".format)
final_summary_most_popular['Total Purchase Value']=final_summary_most_popular['Total Purchase Value'].map("${:.2f}".format)

final_summary_most_popular
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Popular_Items.sort_values('Price', ascending= False)
top5_profit = pd.DataFrame(purchase_df.groupby('Item ID')['Price'].sum())
top5_profit.sort_values('Price', ascending = False, inplace = True)
# only keep top 5
top5_profit = top5_profit.iloc[0:5][:]
#item purchase count
pur_count_profit = pd.DataFrame(purchase_df.groupby('Item ID')['Item ID'].count())

top5_profit = pd.merge(top5_profit, pur_count_profit, left_index = True, right_index = True, how = 'left')
top5_merge_profit = pd.merge(top5_profit, no_dup_items, left_index = True, right_on = 'Item ID', how = 'left')
top5_merge_profit = top5_merge_profit[['Item ID', 'Item Name', 'Item ID_x', 'Price_y','Price_x']]
top5_merge_profit.set_index(['Item ID'], inplace=True)
top5_merge_profit.rename(columns = {'Item ID_x': 'Purchase Count', 'Price_y': 'Item Price', 'Price_x': 'Total Purchase Value'}, inplace = True)

top5_merge_profit
```
